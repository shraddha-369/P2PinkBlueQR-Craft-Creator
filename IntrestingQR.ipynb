```python
import qrcode
from PIL import Image
from IPython.display import display

```

```python
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=5,
    border=10,
)


qr.add_data("https://www.linkedin.com/in/shraddha-rawat-0b47aa309/")
qr.make(fit=True)


img = qr.make_image(fill_color="blue", back_color="pink")

# Show the image in Colab (no saving needed)
display(img)

```
